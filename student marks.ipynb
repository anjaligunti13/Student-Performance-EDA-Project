```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\anjal\Downloads\StudentsPerformance.csv")
```

```python
data
```

```python
#basic checks
data.info()
data.describe()
```

```python
#checking null values
data.isnull().sum()
```

```python
#score distribution -overall perfomance
plt.figure(figsize=(10,4))
sns.histplot(data['math score'],kde=True)
plt.title('Math score distribution',fontsize=18)
plt.xlabel("Math score",fontsize=16)
plt.ylabel("count",fontsize=16)
plt.show()

plt.figure(figsize=(10,4))
sns.histplot(data['reading score'],kde=True,color='orange')
plt.title('reading score distribution',fontsize=18)
plt.xlabel("reading score",fontsize=16)
plt.ylabel("count",fontsize=16)
plt.show()

plt.figure(figsize=(10,4))
sns.histplot(data['writing score'],kde=True,color='green')
plt.title('writing score distribution',fontsize=18)
plt.xlabel("writing score",fontsize=16)
plt.ylabel("count",fontsize=16)
plt.show()
```

```python
#Gender vs Average Scores
#math score average
plt.figure(figsize=(5,5))
sns.barplot(x='gender',y='math score',data=data,width=0.2,edgecolor='black')
plt.title('Math score vs gender',fontsize=18,color='red')
plt.xlabel("Gender",fontsize=16,color='navy')
plt.ylabel("count",fontsize=16,color='navy')
plt.show()

#reading score average
plt.figure(figsize=(5,5))
sns.barplot(x='gender',y='reading score',data=data,width=0.2,edgecolor='black')
plt.title('reading score vs gender',fontsize=18,color='red')
plt.xlabel("Gender",fontsize=16,color='navy')
plt.ylabel("count",fontsize=16,color='navy')
plt.show()

#writing score average
plt.figure(figsize=(5,5))
sns.barplot(x='gender',y='writing score',data=data,width=0.2,edgecolor='black')
plt.title('writing score vs gender',fontsize=18,color='red')
plt.xlabel("Gender",fontsize=16,color='navy')
plt.ylabel("count",fontsize=16,color='navy')
plt.show()
```

```python
#Parental Education vs Scores
plt.figure(figsize=(5,5))
sns.boxplot(x='parental level of education',y='math score',data=data,width=0.2)
plt.title('math score vs gender',fontsize=18,color='red')
plt.xlabel("Gender",fontsize=16,color='navy')
plt.ylabel("count",fontsize=16,color='navy')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(5,5))
sns.boxplot(x='parental level of education',y='reading score',data=data,width=0.2)
plt.title('reading score vs gender',fontsize=18,color='red')
plt.xlabel("Gender",fontsize=16,color='navy')
plt.ylabel("count",fontsize=16,color='navy')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(5,5))
sns.boxplot(x='parental level of education',y='writing score',data=data,width=0.2)
plt.title('writing score vs gender',fontsize=18,color='red')
plt.xlabel("Gender",fontsize=16,color='navy')
plt.ylabel("count",fontsize=16,color='navy')
plt.xticks(rotation=90)
plt.show()
```

```python
#Score Correlation Heatmap
plt.figure(figsize=(6,7))
sns.heatmap(data[['math score','reading score','writing score']].corr(),annot=True,cmap='Blues')
plt.title("corelation of heat map")
plt.show()
```

```python
# Test Preparation Impact
plt.figure(figsize=(6,7))
sns.barplot(x='test preparation course',y='math score',data=data,width=0.2)
plt.title("Test preparation vs Math score",color='navy',fontsize=18)
plt.show()
```

```python
#Race/Ethnicity Comparison
plt.figure(figsize=(10,5))
sns.barplot(x='race/ethnicity', y='math score', data=data,width=0.2)
plt.title('Race/Ethnicity vs Math Score',fontsize=20)
plt.xlabel("race/ethnicity",color='red',fontsize=18)
plt.ylabel("math score",color='red',fontsize=18)
plt.show()

plt.figure(figsize=(10,5))
sns.barplot(x='race/ethnicity', y='reading score', data=data,width=0.2)
plt.title('Race/Ethnicity vs reading Score',fontsize=20)
plt.xlabel("race/ethnicity",color='red',fontsize=18)
plt.ylabel("reading score",color='red',fontsize=18)
plt.show()

plt.figure(figsize=(10,5))
sns.barplot(x='race/ethnicity', y='writing score', data=data,width=0.2)
plt.title('Race/Ethnicity vs writing Score',fontsize=20)
plt.xlabel("race/ethnicity",color='red',fontsize=18)
plt.ylabel("writing score",color='red',fontsize=18)
plt.show()
```

```python
The data suggests that gender and test preparation have a notable influence on students' academic performance. Girls tend to perform better in reading and writing, while boys perform slightly better in math. Students who completed the test preparation course show improved math scores, indicating the effectiveness of additional support. Race/ethnicity does not significantly affect performance, suggesting equal learning opportunities across groups. Overall, most students score between 60 and 90, reflecting a generally strong academic level.
```
